```python
import numpy as np
import matplotlib.pyplot as plt

data = {k: 0 for k in range(0, 7500)}

# read communities from file and count their number of entries
with open('ig_communities_output.metis', 'r') as inputFile:
    for line in inputFile:
        splittedLine = line.split(' ')
        data[len(splittedLine)] += 1

print(data)
```

```python
fig = plt.figure(figsize = (10, 5))

plt.stairs(data.values(),  range(0, 7501))
plt.yscale('log')
plt.xscale('log')

plt.ylabel("# of communities")
plt.xlabel("# of members in community")

plt.show()
```
